### Collect Investments Data Using Alpaca: `SPY` (stocks) and `AGG` (bonds)

```python
# Set current amount of shares
my_agg = 200
my_spy = 50
```

```python
etf_assets = {"shares": [my_agg, my_spy]}

# Set the tickers
tickers = ["AGG", "SPY"]

# Create the shares DataFrame with `tickers` as index
df_etf = pd.DataFrame(etf_assets, index=tickers)

# Display shares data
df_etf
```

```python
# Set Alpaca API key and secret
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")
```

```python
alpaca_api_key
```

```python
# Create the Alpaca API object
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
# Format current date as ISO format
# I obtained this code from: https://www.tutorialspoint.com/get-the-most-recent-previous-business-day-using-python
# to get the latest available business day to retrieve data
def get_previous_business_day():
    today = pd.Timestamp.today()
    previous_day = today - pd.DateOffset(days=1)
    previous_business_day = pd.bdate_range(end=previous_day, periods=1)[0]
    return previous_business_day.strftime("%Y-%m-%d")
```

```python
today = get_previous_business_day()
today
```

```python
# Set timeframe to "1Day" for Alpaca API
timeframe = "1Day"
```

```python
# Get current closing prices for SPY and AGG
etf_portfolio = alpaca.get_bars(
    tickers,
    timeframe,
    start = today,
    end = today
).df
```

```python
etf_portfolio
```

```python
# Reorganize the DataFrame
# Separate ticker data
AGG = etf_portfolio[etf_portfolio['symbol']=='AGG'].drop('symbol', axis=1)
SPY = etf_portfolio[etf_portfolio['symbol']=='SPY'].drop('symbol', axis=1)
```

```python
SPY
```

```python
# Concatenate the ticker DataFrames
df_etf_portfolio = pd.concat([AGG, SPY],axis=1, keys=['AGG','SPY'])
# Preview DataFrame
df_etf_portfolio
```

```python
# Pick AGG and SPY close prices
agg_close_price = float(df_etf_portfolio["AGG"]["close"])
spy_close_price = float(df_etf_portfolio["SPY"]["close"])
```

```python
# Print AGG and SPY close prices
print(f"Current AGG closing price: ${agg_close_price}")
print(f"Current SPY closing price: ${spy_close_price}")
```

```python
agg_value = agg_close_price * df_etf.loc["AGG"]["shares"]
spy_value = spy_close_price * df_etf.loc["SPY"]["shares"]
value_of_etf = agg_value + spy_value
```

## Part 2 - Retirement Planning

### Monte Carlo Simulation

```python
# Set start and end dates of five years back from today.
# Sample results may vary from the solution based on the time frame chosen
start_date = pd.Timestamp('2016-05-01', tz='America/New_York').isoformat()
end_date = pd.Timestamp('2021-05-01', tz='America/New_York').isoformat()
```

```python
# Get 5 years' worth of historical data for SPY and AGG
etf_portfolio_MC = alpaca.get_bars(
    tickers,
    timeframe,
    start = start_date,
    end = end_date
).df
etf_portfolio_MC.head()
```

```python
# Reorganize the DataFrame
# Separate ticker data
AGG_MC = etf_portfolio_MC[etf_portfolio_MC['symbol']=='AGG'].drop(['symbol', 'vwap', 'trade_count'], axis=1)
SPY_MC = etf_portfolio_MC[etf_portfolio_MC['symbol']=='SPY'].drop(['symbol', 'vwap', 'trade_count'], axis=1)
```

```python
AGG_MC.head()
```

```python
SPY_MC.head()
```

```python
# Concatenate the ticker DataFrames
df_etf_portfolio_MC = pd.concat([AGG_MC, SPY_MC],axis=1, keys=['AGG','SPY'])
# Display sample data
df_etf_portfolio_MC.head()
```

```python
# Configuring a Monte Carlo simulation to forecast 30 years cumulative returns
MC_30_years = MCSimulation(
    portfolio_data = df_etf_portfolio_MC,
    weights = [.40,.60],
    num_simulation = 500,
    num_trading_days = 252*30
)
```

```python
MC_30_years.portfolio_data.head()
```

```python
MC_30_years.portfolio_data.tail()
```

```python
MC_30_years.calc_cumulative_return()
```

```python
# Plot simulation outcomes
line_plot_MC = MC_30_years.plot_simulation()
```

```python
# Plot probability distribution and confidence intervals
dist_plot_MC = MC_30_years.plot_distribution()
```
